```python
import requests
import json

url = "http://127.0.0.1:8000/mcp"  # Replace with your MCP server URL

payload = {
    "jsonrpc": "2.0",
    "method": "duckduckgo_search",
    "params": {"query": "Python MCP SDK"},
    "id": 1
}

headers = {
    "Content-Type": "application/json",
    "Accept": "application/json"
}

response = requests.post(url, headers=headers, json=payload)
print(json.dumps(response.json(), indent=2))
```

```python
import requests
import json

url = "http://127.0.0.1:8000/mcp"

# Step 1: create session
session_payload = {
    "jsonrpc": "2.0",
    "method": "session.create",
    "params": {},
    "id": 1
}
response = requests.post(url, json=session_payload)
session_id = response.json()["result"]["session_id"]

# Step 2: call tool with session
tool_payload = {
    "jsonrpc": "2.0",
    "method": "duckduckgo_search",
    "params": {"session_id": session_id, "query": "Python MCP SDK"},
    "id": 2
}

response = requests.post(url, json=tool_payload)
print(json.dumps(response.json(), indent=2))
```

```python
import json
import requests

# URL of the running MCP server
MCP_URL = "http://127.0.0.1:5003"

def call_duckduckgo_search(query: str):
    payload = {
        "tool": "duckduckgo_search",
        "input": query
    }

    # Send a POST request
    try:
        response = requests.post(MCP_URL, json=payload, timeout=10)
        response.raise_for_status()
        data = response.json()
        return data
    except Exception as e:
        print("Error:", e)
        return None

if __name__ == "__main__":
    results = call_duckduckgo_search("Python programming")
    print(json.dumps(results, indent=2))
```
